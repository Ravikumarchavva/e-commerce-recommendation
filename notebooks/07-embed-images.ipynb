```python
import polars as pl
import psycopg2
from psycopg2.extras import execute_batch
from e_commerce_recommendation.configs.settings import Settings

def insert_products(df, db_url):
    conn = psycopg2.connect(db_url)
    cur = conn.cursor()

    rows = [
        (
            r["asin"], r["country"], r["title"],
            r["image_path"], r["product_url"],
            r["stars"], r["reviews"], r["price"],
            r["is_best_seller"], r["bought_in_last_month"],
            [r["labels"]]
        )
        for r in df.to_dicts()
    ]

    execute_batch(
        cur,
        """
        INSERT INTO products
        (asin, country, title, image_path, product_url, stars, reviews,
         price, is_best_seller, bought_in_last_month, labels)
        VALUES (%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s)
        ON CONFLICT (asin, country) DO NOTHING
        """,
        rows,
        page_size=50000
    )

    conn.commit()
    cur.close()
    conn.close()

settings = Settings()
insert_products(df, settings.database_url)
```

```python
from vllm import LLM

llm = LLM(
    model="Qwen/Qwen3-VL-Embedding-2B",
    runner="pooling",
    max_model_len=512,   # correct for images
    quantization="mxfp4",
    gpu_memory_utilization=0.85
)
```

```python
from PIL import Image

def build_multimodal_prompt(image: Image.Image, title: str):
    return [
        {
            "role": "system",
            "content": "You are a multimodal embedding model. Encode the visual appearance and the product title into one semantic vector."
        },
        {
            "role": "user",
            "content": [
                {"type": "image", "image": image},
                {"type": "text", "text": f"Product title: {title}"}
            ]
        }
    ]

def embed_product(image: Image.Image, title: str, llm: LLM):
    conversation = build_multimodal_prompt(image, title)

    vllm_input = llm.llm_engine.tokenizer.apply_chat_template(
        conversation,
        tokenize=False,
        add_generation_prompt=True
    )

    out = llm.embed([vllm_input])[0]
    return out.outputs.embedding

def embed_products_batch(images, titles, llm: LLM):
    inputs = []

    for img, title in zip(images, titles):
        conversation = build_multimodal_prompt(img, title)
        prompt = llm.llm_engine.tokenizer.apply_chat_template(
            conversation,
            tokenize=False,
            add_generation_prompt=True
        )
        inputs.append(prompt)

    outputs = llm.embed(inputs)
    return [o.outputs.embedding for o in outputs]
```

```python
import asyncpg
from PIL import Image
from vllm import LLM
from tqdm.notebook import tqdm
from e_commerce_recommendation.configs.settings import Settings

settings = Settings()
BATCH_SIZE = 9
DB_BATCH = 10
DATABASE_URL = settings.DATABASE_URL
COUNT_SQL = "SELECT COUNT(*) FROM products WHERE embedding IS NULL"
FETCH_SQL = """
SELECT id, image_path, title
FROM products
WHERE embedding IS NULL
LIMIT $1
"""
def to_pgvector(vec):
    return "[" + ",".join(f"{x:.6f}" for x in vec) + "]"

async def embed_all_products(pool: asyncpg.Pool, llm: LLM):

    async with pool.acquire() as conn:
        total = await conn.fetchval(COUNT_SQL)

    if total == 0:
        print("Nothing to embed.")
        return

    print(f"Embedding {total:,} products")

    pbar = tqdm(total=total, desc="Embedding", unit="img")

    while True:
        async with pool.acquire() as conn:
            rows = await conn.fetch(FETCH_SQL, DB_BATCH)

        if not rows:
            break

        ids, images, titles = [], [], []

        for r in rows:
            try:
                img = Image.open(r["image_path"]).convert("RGB")
                images.append(img)
                titles.append(r["title"])
                ids.append(r["id"])
            except:
                continue

        for i in range(0, len(images), BATCH_SIZE):
            batch_imgs = images[i:i+BATCH_SIZE]
            batch_titles = titles[i:i+BATCH_SIZE]
            batch_ids = ids[i:i+BATCH_SIZE]

            vectors = embed_products_batch(batch_imgs, batch_titles, llm)

            async with pool.acquire() as conn:
                async with conn.transaction():
                    await conn.executemany(
                        "UPDATE products SET embedding=$2::vector WHERE id=$1",
                        [(pid, to_pgvector(vec)) for pid, vec in zip(batch_ids, vectors)]
                    )

            pbar.update(len(batch_ids))

    pbar.close()
    print("All products embedded.")

    
pool = await asyncpg.create_pool(DATABASE_URL)

await embed_all_products(pool, llm)
```
