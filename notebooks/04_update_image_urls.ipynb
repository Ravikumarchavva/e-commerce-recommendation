```python
from e_commerce_recommendation.configs.settings import settings
import polars as pl

DATA_DIR = settings.BASE_DIR / "data"
RAW_DATA_DIR = DATA_DIR / "raw_data"
PROCESSED_DATA_DIR = DATA_DIR / "processed_data"

products = pl.read_parquet(PROCESSED_DATA_DIR / "products_with_canonical_categories.parquet")
products
```

```python
india_images_dir = RAW_DATA_DIR / "india/images"
canada_images_dir = RAW_DATA_DIR / "canada/images"
uk_images_dir = RAW_DATA_DIR / "uk/images"
usa_images_dir = RAW_DATA_DIR / "usa/images"
# products['imgUrl'][0].split('/')[-1]
def update_image_url(row):
    img_filename = row['asin']
    country = row['country']
    if country == 'india':
        new_url = f"{india_images_dir}/{img_filename}.jpg"
    elif country == 'canada':
        new_url = f"{canada_images_dir}/{img_filename}.jpg"
    elif country == 'uk':
        new_url = f"{uk_images_dir}/{img_filename}.jpg"
    elif country == 'usa':
        new_url = f"{usa_images_dir}/{img_filename}.jpg"
    else:
        new_url = row['imgUrl']  # Fallback to original URL if country is unknown
    return new_url

products = products.with_columns(
    pl.struct(['asin', 'imgUrl', 'country']).map_elements(update_image_url, return_dtype=pl.String).alias('imgUrl')
)
products
```

```python
products = products.rename({
    "asin": "asin",
    "imgUrl": "image_path",
    "productURL": "product_url",
    "isBestSeller": "is_best_seller",
    "cluster_id": "labels",
    "boughtInLastMonth": "bought_in_last_month",
}).select([
    "asin",
    "country",
    "title",
    "image_path",
    "product_url",
    "stars",
    "reviews",
    "price",
    "is_best_seller",
    "bought_in_last_month",
    "labels",
])
products
```

```python
from PIL import Image
from concurrent.futures import ThreadPoolExecutor, as_completed
import os
from tqdm import tqdm

def is_valid_image(image_path):
    """Validate single image - returns True/False."""
    if not os.path.exists(image_path):
        return False
    try:
        with Image.open(image_path) as img:
            img.verify()
            if img.size[0] < 50 or img.size[1] < 50:
                return False
        return True
    except (IOError, SyntaxError):
        return False
    
products_titile_filtered = (
    products
    .filter(pl.col("title").str.len_chars() > 5)
)

print(f"Pre-filtered: {len(products):,} → {len(products_titile_filtered):,} products")
```

```python
# Extract image paths to list
image_paths = products_titile_filtered["image_path"].to_list()

# Concurrent validation with ThreadPoolExecutor
validation_results = [False] * len(image_paths)  # Initialize results list

print(f"Validating {len(image_paths):,} images concurrently...")

with ThreadPoolExecutor(max_workers=128) as executor:
    # Submit all tasks
    future_to_idx = {
        executor.submit(is_valid_image, path): idx 
        for idx, path in enumerate(image_paths)
    }
    
    # Collect results with progress bar
    for future in tqdm(as_completed(future_to_idx), total=len(image_paths)):
        idx = future_to_idx[future]
        try:
            validation_results[idx] = future.result()
        except Exception:
            validation_results[idx] = False
```

```python
# Add validation column and filter
products_filtered = (
    products_titile_filtered
    .with_columns(pl.Series("is_valid_image", validation_results))
    .filter(pl.col("is_valid_image"))
    .drop("is_valid_image")  # Remove helper column
)

print(f"Final: {len(products_filtered):,} valid products")
products_filtered
```

```python
products_filtered =products_filtered.with_columns(pl.col("labels").cast(pl.List(pl.Int64)).cast(pl.String))
products_filtered
```

```python
products_filtered.write_parquet(PROCESSED_DATA_DIR / "products_final.parquet")
```

```python
import polars as pl
import psycopg2
from psycopg2.extras import execute_batch
from e_commerce_recommendation.configs.settings import Settings

def insert_products(df, db_url):
    conn = psycopg2.connect(db_url)
    cur = conn.cursor()

    rows = [
        (
            r["asin"], r["country"], r["title"],
            r["image_path"], r["product_url"],
            r["stars"], r["reviews"], r["price"],
            r["is_best_seller"], r["bought_in_last_month"],
            [r["labels"]]
        )
        for r in df.to_dicts()
    ]

    execute_batch(
        cur,
        """
        INSERT INTO products
        (asin, country, title, image_path, product_url, stars, reviews,
         price, is_best_seller, bought_in_last_month, labels)
        VALUES (%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s)
        ON CONFLICT (asin, country) DO NOTHING
        """,
        rows,
        page_size=50000
    )

    conn.commit()
    cur.close()
    conn.close()

settings = Settings()
insert_products(products, settings.DATABASE_URL)
```
