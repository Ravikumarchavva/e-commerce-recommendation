```python
import os
import asyncio
import aiohttp
from tqdm.asyncio import tqdm
import polars as pl
from typing import Optional


async def download_image(
    session: aiohttp.ClientSession,
    row: dict,
    download_path: str,
    semaphore: asyncio.Semaphore,
    max_retries: int = 3
) -> tuple[bool, Optional[str]]:
    """Download a single image with retry logic and error handling."""
    image_url = row["imgUrl"]
    image_id = row["asin"]
    file_path = os.path.join(download_path, f"{image_id}.jpg")

    # Skip if already exists
    if os.path.exists(file_path):
        return True, None

    # Move semaphore outside retry loop to fix concurrency bottleneck
    for attempt in range(max_retries):
        try:
            async with semaphore:  # Only hold semaphore during actual request
                async with session.get(image_url, timeout=10) as response:
                    if response.status == 200:
                        content = await response.read()
                        
                        # Write atomically using temp file
                        temp_path = f"{file_path}.tmp"
                        with open(temp_path, "wb") as f:
                            f.write(content)
                        os.rename(temp_path, file_path)
                        
                        return True, None
                    elif response.status == 404:
                        return False, f"404 Not Found: {image_id}"
                    else:
                        # Don't retry on client errors (4xx)
                        if 400 <= response.status < 500:
                            return False, f"HTTP {response.status}: {image_id}"
                        # Retry on server errors (5xx)
                        if attempt < max_retries - 1:
                            await asyncio.sleep(2 ** attempt)
                            continue
                        return False, f"HTTP {response.status}: {image_id}"
                        
        except asyncio.TimeoutError:
            if attempt < max_retries - 1:
                await asyncio.sleep(2 ** attempt)
                continue
            return False, f"Timeout: {image_id}"
            
        except Exception as e:
            if attempt < max_retries - 1:
                await asyncio.sleep(2 ** attempt)
                continue
            return False, f"Error ({type(e).__name__}): {image_id}"
    
    return False, f"Max retries exceeded: {image_id}"


async def download_all_images(
    df: pl.DataFrame,
    download_path: str,
    max_concurrent: int = 100,
    max_retries: int = 3
) -> dict:
    """Download all images with progress tracking and error reporting."""
    os.makedirs(download_path, exist_ok=True)
    
    rows = df.to_dicts()
    semaphore = asyncio.Semaphore(max_concurrent)
    
    # Simpler connector config matching threading behavior
    connector = aiohttp.TCPConnector(
        limit=0,  # No limit on total connections
        limit_per_host=0,  # No per-host limit
        ttl_dns_cache=300
    )
    
    async with aiohttp.ClientSession(connector=connector) as session:
        tasks = [
            download_image(session, row, download_path, semaphore, max_retries)
            for row in rows
        ]
        
        results = await tqdm.gather(*tasks, desc="Downloading images")
    
    # Analyze results
    successes = sum(1 for success, _ in results if success)
    failures = [(error) for success, error in results if not success and error]
    
    stats = {
        "total": len(results),
        "success": successes,
        "failed": len(failures),
        "success_rate": f"{(successes/len(results)*100):.1f}%"
    }
    
    print(f"\nüìä Download Statistics:")
    print(f"  Total: {stats['total']}")
    print(f"  Success: {stats['success']} ({stats['success_rate']})")
    print(f"  Failed: {stats['failed']}")
    
    if failures[:10]:  # Show first 10 errors
        print(f"\n‚ö†Ô∏è  Sample errors:")
        for error in failures[:10]:
            print(f"  - {error}")
    
    return stats


download_path = f"{BASE_DIR / 'data/raw_data/india/images'}"
# stats = await download_all_images(df, download_path, max_concurrent=200, max_retries=3)
```
